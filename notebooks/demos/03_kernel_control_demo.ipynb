## 1. Running Cells Sequentially

This section demonstrates how the kernel remembers variables across cells.

```python
x = 10
print("x is:", x)
```

```python
x = x + 5
print("Updated x:", x)
```

```python
print("Final x:", x)
```

## 3. Restarting the Kernel

After restart, variables are cleared from memory.

```python
print(x)
```

## 4. Restart vs Interrupt

- Interrupt stops the currently running cell without clearing variables.
- Restart clears the entire kernel memory including variables and imports.
- Restart ensures reproducibility and prevents hidden state issues.
