## When to Use Code vs Markdown

### Code Cells
Used to:
- Execute Python logic
- Perform calculations
- Train models
- Process data

### Markdown Cells
Used to:
- Explain why something is done
- Interpret outputs
- Structure the notebook
- Add headings and documentation
